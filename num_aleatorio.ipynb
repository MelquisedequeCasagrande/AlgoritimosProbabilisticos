```python

import random
from math import sqrt, log
import numpy as np
import matplotlib.pyplot as plt

def SPTINIK(m, s):   
  p0 = 0.322232431088   
  q0 = 0.099348462606
  p1 = 1.0
  q1 = 0.588581570495
  p2 = 0.342242088547
  q2 = 0.531103462366
  p3 = 0.204231210245e-1
  q3 = 0.103537752850
  p4 = 0.453642210148e-4
  q4 = 0.385607006340e-2

  u   = random.random() # Retorna um número aleatório entre 0.0 e 1.0

  if (u < 0.5):
    t = sqrt(-2.0 * log(u))

  else:
    t = sqrt(-2.0 * log(1.0 - u))

  p   = p0 + t * (p1 + t * (p2 + t * (p3 + t * p4)))
  q   = q0 + t * (q1 + t * (q2 + t * (q3 + t * q4)))

  if (u < 0.5):
    z = (p / q) - t

  else:
    z = t - (p / q)

  return (m + s * z)

def myrand():
    return random.random()


aleat1 = []
aleat2 = []
def main():
    for i in range(100000):
        aleat1.append(myrand())
        aleat2.append(SPTINIK(3, 0.1))
main()

plt.hist(aleat2, bins= 1000)
plt.show()
```
