```python
#5 Correlação entre primeiro sorteio e o segundo sorteio, entre o primeiro e o terceiro.....
import numpy as np
import pandas as pd

# Transpor o DataFrame
teste = data.T

# Selecionando o 1° sorteio
linha1 = pd.to_numeric(teste.iloc[2:17, 2719], errors='coerce').to_numpy()

# Remover valores NaN
linha1 = linha1[~np.isnan(linha1)]

segundo_sorteio = 2

for x in range(len(data) - 2, -1, -1):
  linha2 = pd.to_numeric(teste.iloc[2:17, x], errors='coerce').to_numpy()
  linha2 = linha2[~np.isnan(linha2)]

  # Calcular a correlação
  correlacao = np.corrcoef(linha1, linha2)[0, 1]
  print("Correlação 1° Sorteio e " + str(segundo_sorteio) + "° Sorteio: " + str(correlacao))
  segundo_sorteio = segundo_sorteio + 1
```

```python
#7.4 Variância por sorteio
for dia_semana, dados_dia_semana in data.groupby(data.index.weekday):
  print(f'Dia da semana: {nomes_semana[dia_semana]}')
  print(dados_dia_semana.iloc[0:, 18:])
  print()
```

```python
#7.5 Correlação entre sorteio o primeiro sorteio de segunda, com o primeiro sorteio de quarta...
import numpy as np
import pandas as pd

for dia_semana, dados_dia_semana in data.groupby(data.index.weekday):

  if(dia_semana == 0): # Filtrando dataframe de segunda

    # Selecionando o 1° sorteio de segunda
    segunda = dados_dia_semana.tail(n=1)
    segunda = pd.to_numeric(segunda.iloc[0,1:16], errors='coerce').to_numpy()
    # Remover valores NaN
    segunda = segunda[~np.isnan(segunda)]

  if(dia_semana != 0):

    # Selecionando os primeiros sorteios das outras semanas
    linha2 = dados_dia_semana.tail(n=1)

    # Selecionando os primeiros sorteios das outras semanas
    linha2 = pd.to_numeric(linha2.iloc[0,1:16], errors='coerce').to_numpy()
    linha2 = linha2[~np.isnan(linha2)]

    # Calcular a correlação
    correlacao = np.corrcoef(segunda, linha2)[0, 1]
    print("Correlação do 1° Sorteio de Segunda-feira e 1° Sorteio de " + str(nomes_semana[dia_semana]) + " = " + str(correlacao))
```

```python
#7.6 Covariância entre primeiro sorteio sorteio de segunda, com o primeiro sorteio de quarta...
import numpy as np
import pandas as pd

for dia_semana, dados_dia_semana in data.groupby(data.index.weekday):

  if(dia_semana == 0): # Filtrando dataframe de segunda

    # Selecionando o 1° sorteio de segunda
    segunda = dados_dia_semana.tail(n=1)
    segunda = pd.to_numeric(segunda.iloc[0,1:16], errors='coerce').to_numpy()
    # Remover valores NaN
    segunda = segunda[~np.isnan(segunda)]

  if(dia_semana != 0):

    # Selecionando os primeiros sorteios das outras semanas
    linha2 = dados_dia_semana.tail(n=1)

    # Selecionando os primeiros sorteios das outras semanas
    linha2 = pd.to_numeric(linha2.iloc[0,1:16], errors='coerce').to_numpy()
    linha2 = linha2[~np.isnan(linha2)]

    # Calcular a Covariância
    covariancia = np.cov(segunda, linha2)[0, 1]
    print("Covariância do 1° Sorteio de Segunda-feira e 1° Sorteio de " + str(nomes_semana[dia_semana]) + " = " + str(covariancia))
```

```python
#8 Separar a base de dados por meses, e verificar se o padrão de sorteio se repete dentro do ano
# Separar o DataFrame por ano usando o método groupby e pd.Grouper
anos = data.groupby(pd.Grouper(freq='Y'))

# Iterar sobre os anos
for ano, dados_ano in anos:

  for x in range(len(dados_ano)):
    # Selecionando linha para comparar com outros dias nos outros meses
    linha_1 = pd.to_numeric(dados_ano.iloc[x,1:16], errors='coerce').to_numpy()
    # Remover valores NaN
    linha_1 = linha_1[~np.isnan(linha_1)]
    #print("Linha 1 = " + str(linha_1))

    for ano_2, dados_ano_2 in anos:

      if(ano.year == ano_2.year):
        for i in range(len(dados_ano_2)):
          # Selecionando linha para comparar com outros dias nos outros meses
          linha_2 = pd.to_numeric(dados_ano_2.iloc[i,1:16], errors='coerce').to_numpy()
          # Remover valores NaN
          linha_2 = linha_2[~np.isnan(linha_2)]
          #print("Linha 2 = " + str(linha_2))
          if np.array_equal(linha_2, linha_1):
            if(dados_ano.iloc[x].name != dados_ano_2.iloc[i].name):
              print("Ano: " + str(ano.year) + " Sorteiro: " + str(dados_ano.iloc[x,0]) + " e Sorteiro: " + str(dados_ano_2.iloc[i,0]) + " os valores são iguais.")
  print(str(ano.year) + " Verificado")
```

```python
#9 Verificar se o padrão se repete dentro do mês
# Separar o DataFrame por mês usando o método groupby e pd.Grouper
meses = data.groupby(pd.Grouper(freq='M'))

# Iterar sobre os meses
for mes, dados_mes in meses:

  for x in range(len(dados_mes)):
    # Selecionando linha para comparar com outros dias nos outros meses
    linha_1 = pd.to_numeric(dados_mes.iloc[x,1:16], errors='coerce').to_numpy()
    # Remover valores NaN
    linha_1 = linha_1[~np.isnan(linha_1)]
    #print("Linha 1 = " + str(linha_1))

    for mes_2, dados_mes_2 in meses:

      if(mes.month == mes_2.month):
        for i in range(len(dados_mes_2)):
          # Selecionando linha para comparar com outros dias nos outros meses
          linha_2 = pd.to_numeric(dados_mes_2.iloc[i,1:16], errors='coerce').to_numpy()
          # Remover valores NaN
          linha_2 = linha_2[~np.isnan(linha_2)]
          #print("Linha 2 = " + str(linha_2))
          if np.array_equal(linha_2, linha_1):
            if(dados_mes.iloc[x].name != dados_mes_2.iloc[i].name):
              print("Mês: " + str(mes) + " Sorteiro: " + str(dados_mes.iloc[x,0]) + " e Sorteiro: " + str(dados_mes_2.iloc[i,0]) + " os valores são iguais.")
  print(str(mes) + " Verificado")
```
