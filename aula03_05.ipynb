```python
#3

import numpy as np
import matplotlib.pyplot as plt

# Covarincia
def cov(x, y):
  xbar, ybar = x.mean(), y.mean()
  return np.sum((x - xbar)*(y - ybar))/(len(x) - 1)

# Matriz de covarincia
def cov_mat(X):
  return np.array([[cov(X[0], X[0]), cov(X[0], X[1])] , [cov(X[1], X[0]), cov(X[1], X[1])]])

X = X - np.mean(X, 0)

# Matriz de escala
sx, sy = 0.7, 3.4
Scale = np.array([[sx, 0], [0, sy]])
                           
# Aplicando matriz de escala para X
Y = X.dot(Scale)
plt.scatter(Y[:, 0], Y[:, 1])
plt.title('Dados aps transformao linear')
plt.axis('equal')
plt.savefig("ex_2")
 
# Clculo da matriz de covarincia
print(cov_mat(Y.T),"\n")
```

```python
#4

import numpy as np
import matplotlib.pyplot as plt

# Covarincia
def cov(x, y):
  xbar, ybar = x.mean(), y.mean()
  return np.sum((x - xbar)*(y - ybar))/(len(x) - 1)

a = np.array([1,2,3])
b = np.array([1,2,1])

print(np.cov(a,b,fweights=np.array([1,1,1])),"\n")
# Covarincia padro entre os dois arrays, a covariao entre os dois arrays  zero\
# pois no h pesos particulares, havendo apenas uma varincia para o array b

print(np.cov(a,b,fweights=np.array([1,1,0])),"\n")
# Neste clculo, os valores na posio "3" dos arrays so desconsiderados por conta\
# do peso nulo, logo, os dois arrays sero iguais e as varincias, juntamente com a\
# covarincia, tambm sero iguais

print(np.cov(a,b,fweights=np.array([100,100,1])),"\n")
# Neste caso, os valores na posio 3 dos dois arrays ainda possuem 1% de peso\
# comparado aos outros elementos, contudo, a varincia e covarincia se dar muito mais\
# entre os outros valores, logo, os resultados sero perto de 0,25 tendo em vista que\
# os valores dos dois arrays so idnticos

print(np.cov(a,b,fweights=np.array([1,1,100])),"\n")
# Para este caso, os valores que eram iguais para os dois arrays esto com pesos\
# muito baixos, o que ir influenciar na covarincia, e a diferena entre os\
# valores dos dois arrays na posio 3 tambm ir influenciar o resultado da covarincia
```

```python
#6

import numpy as np
import matplotlib.pyplot as plt

# Covarincia
def cov(x, y):
  xbar, ybar = x.mean(), y.mean()
  return np.sum((x - xbar)*(y - ybar))/(len(x) - 1)

a = np.array([1,2,3])

print(cov(a,a))
print(cov(a,2*a))
print(cov(a,3*a))
print(cov(a, -a))
print(cov(a, a**2))
print(cov(a, a**3))

# Um valor positivo de covarincia simboliza um valor "mais linear"
```

```python
#7
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean, stdev

def cov(x, y):
  xbar = x.mean()
  ybar = y.mean()
  return np.sum((x - xbar)*(y - ybar))/(len(x) - 1)

def corr(x, y):
  return cov(x, y)/(np.std(x)*np.std(y))

x = np.random.normal(0, 1, 500)
y = np.random.normal(10, 2, 500)

print(corr(x, y))
print(np.corrcoef(x,y))
```

```python
#8
import numpy as np
import matplotlib.pyplot as plt

def cov(x, y):
  xbar = x.mean()
  ybar = y.mean()
  return np.sum((x - xbar)*(y - ybar))/(len(x) - 1)

def corr(x, y):
  return cov(x, y)/(np.std(x)*np.std(y))

a = np.array([1,2,3])


print("\n")
print(corr(a,a))
print(np.corrcoef(a,a))

print("\n")
print(corr(a,2*a))
print(np.corrcoef(a,2*a))

print("\n")
print(corr(a,3*a))
print(np.corrcoef(a,3*a))

print("\n")
print(corr(a,-a))
print(np.corrcoef(a,-a))

print("\n")
print(corr(a,a**2))
print(np.corrcoef(a,a**2))

print("\n")
print(corr(a,a**2))
print(np.corrcoef(a,a**2))
```
